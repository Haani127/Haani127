```python
sixes=deliveries[deliveries['batsman_runs']==6]
fours=deliveries[deliveries['batsman_runs']==4]
display(sixes.batsman.value_counts().head(9))
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.bar(sixes.batsman.value_counts().head(9).index,sixes.batsman.value_counts().head(9).values)
plt.title("Highest Number of sixes by each batsman")
plt.xlabel('batsman')
plt.ylabel('sixes')
plt.xticks(rotation= 45)
plt.subplot(1,2,2)
plt.bar(fours.batsman.value_counts().head(9).index,fours.batsman.value_counts().head(9).values)
plt.title(" lowest Number of fours by each batsman")
plt.xlabel('batsman')
plt.ylabel('fours')
plt.xticks(rotation= 45)
plt.tight_layout()
plt.show()
```

```python
dismissal=deliveries.dismissal_kind.value_counts()
display(dismissal)
sns.barplot(x=dismissal.index,y=dismissal.values)
plt.title('Dismissal Kinds')
plt.xlabel('types')
plt.ylabel('count')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
most_runs=deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
most_runs=most_runs.sort_index(ascending=True)
display(most_runs)
```

```python
balls_faced=deliveries.batsman.value_counts()
balls_faced=balls_faced.sort_index(ascending=True)
display(balls_faced)
```

```python
balls_faced = balls_faced[balls_faced > 100]
Strike_rate=((most_runs/balls_faced)*100)
display(Strike_rate.sort_values(ascending=False).head(10) .to_frame(name='Strike_rate'))
```
